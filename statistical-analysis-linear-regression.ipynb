# QUESTION#01:

# You're conducting a study to estimate the average daily screen time of students in a specific college. You randomly select 30 students and record their daily screen time. The sample mean screen time is found to be 5.5 hours, with a standard deviation of 1.8 hours. Can this sample provide a reliable estimate of the average daily screen time for all students in the college?


# Part A:  

# Write a Python code to Implement the following steps using:

# i. Import necessary packages.

# ii. Compute and print Standard Error (SE).

# iii. Compute and print upper bound (UB) and lower bound (LB)

```python
# Import necessary packages
import numpy as np
from scipy import stats

# Sample data
sample_mean = 5.5
sample_std_dev = 1.8
sample_size = 30

# Compute Standard Error (SE)
standard_error = sample_std_dev / np.sqrt(sample_size)
print("Standard Error (SE):", standard_error)

# Compute upper bound (UB) and lower bound (LB) for a 95% confidence interval
confidence_level = 0.95
alpha = 1 - confidence_level

t_critical = stats.t.ppf(1 - alpha/2, df=sample_size - 1)
margin_of_error = t_critical * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("Lower Bound (LB):", lower_bound)
print("Upper Bound (UB):", upper_bound)
```

# Part C: 


# Write a Python code to Implement the following steps using: 

# vii. Compute interval using statistics Library.

```python
# Import necessary packages
import statistics
import math

# Sample data
sample_mean = 5.5
sample_std_dev = 1.8
sample_size = 30
confidence_level = 0.95

# Compute the standard error
standard_error = sample_std_dev / math.sqrt(sample_size)

# Compute the margin of error
z_critical = statistics.NormalDist().inv_cdf((1 + confidence_level) / 2)
margin_of_error = z_critical * standard_error

# Compute the confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print("Confidence Interval (95%):")
print("Lower Bound (LB):", lower_bound)
print("Upper Bound (UB):", upper_bound)
```

# QUESTION#02

# You're interested in determining the proportion of residents in a particular town who prefer public transportation for their daily commute. To investigate, you randomly survey 600 residents and discover that 420 of them use public transportation regularly. Can this sample provide a reliable estimate of the proportion of town residents who prefer public transportation?

# Part A: 

# Write a python code to  Implement the following steps.

# i. Import necessary packages.

# ii. Compute and print Standard Error (SE).

# iii. Compute and print upper bound (UB) and lower bound (LB)

```python
# Import necessary packages
import numpy as np
import scipy.stats as stats

# Given values
n = 600
x = 420

# Compute proportion of public transport users
p = x / n

# Compute Standard Error (SE)
SE = np.sqrt(p * (1 - p) / n)

# Compute z-score for 95% confidence interval 
z = stats.norm.ppf(0.975) 

# Compute margin of error
margin_of_error = z * SE

# Compute upper bound (UB) and lower bound (LB) for the confidence interval
UB = p + margin_of_error
LB = p - margin_of_error

print("Standard Error (SE):", SE)
print("Upper Bound (UB):", UB)
print("Lower Bound (LB):", LB)
```

# Part C: 

# Write a python code to  Implement the following steps.

# vii. Compute interval using statistics Library.

```python
# Import necessary packages
import statistics
import math

#Given Data
n = 600
x = 420

# Find Proportion 
p = x / n
# Standard error calculation
standard_error = math.sqrt(p * (1 - p) / n)

# Z-score for 95% confidence interval
z_score = 1.96

# Margin of error calculation
margin_of_error = z_score * standard_error

# Confidence interval calculation
lower_bound = p - margin_of_error
upper_bound = p + margin_of_error

print(f"Proportion of residents who prefer public transportation: {p:.2f}")
print(f"Confidence interval: [{lower_bound:.2f}, {upper_bound:.2f}]")
```

# Tasks I:

# i. Load dataset (Real_Estate_Price.csv) and print first five rows.

```python
import pandas as pd

# Load the dataset
excel_path = 'Real_Estate_Price.xlsx'
df = pd.read_excel(excel_path, sheet_name=0)

# Display the first five rows of the dataset
df.head()
```

# ii. Split the data set into training and test data 75% and 25% respectively.

```python
from sklearn.model_selection import train_test_split

# Select only the relevant columns
df = df[['LotArea', 'SalePrice']]

# Split the data into training (75%) and testing (25%) sets
train, test = train_test_split(df, test_size=0.25, random_state=42)

# Display the number of rows in training and test sets
len_train = len(train)
len_test = len(test)
# (train.head(), len_train, test.head(), len_test)
print(train.head())
```

# iii. Draw feature space (Scatter Plot)

```python
import matplotlib.pyplot as plt

# Plot the scatter plot of LotArea vs. SalePrice
plt.figure(figsize=(10, 6))
plt.scatter(train['LotArea'], train['SalePrice'], color='blue', alpha=0.5, label='Training Data')
plt.scatter(test['LotArea'], test['SalePrice'], color='red', alpha=0.5, label='Test Data')
plt.xlabel('LotArea')
plt.ylabel('SalePrice')
plt.title('Feature Space: LotArea vs. SalePrice')
plt.legend()
plt.grid(True)
plt.show()
```

# Tasks II:

# I. Normalize data 

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler and fit on the training data
scaler = StandardScaler()
train[['LotArea']] = scaler.fit_transform(train[['LotArea']])
test[['LotArea']] = scaler.transform(test[['LotArea']])

# Display first five rows after normalization
train.head()
```

# II. Implement prediction function.

```python
import numpy as np

# Define the prediction function
def predict(X, beta):
    return np.dot(X, beta)

# Add a column of ones for the intercept (beta_0)
train_X = np.c_[np.ones((train.shape[0], 1)), train['LotArea']]
test_X = np.c_[np.ones((test.shape[0], 1)), test['LotArea']]

# Initialize coefficients (beta)
beta_initial = np.zeros(train_X.shape[1])

# Example prediction with initial beta
example_prediction = predict(train_X[:5], beta_initial)
example_prediction
```

# III. Implement Risk function.

```python
# Define the risk function (mean squared error)
def compute_cost(X, y, beta):
    predictions = predict(X, beta)
    errors = predictions - y
    return (1 / (2 * len(y))) * np.dot(errors.T, errors)

# Remove NaN values from the SalePrice column
train_clean = train.dropna(subset=['SalePrice'])

# Extract the corresponding clean training data
train_X_clean = np.c_[np.ones((train_clean.shape[0], 1)), train_clean['LotArea']]
train_y_clean = train_clean['SalePrice'].values

# Example cost calculation with initial beta
initial_cost = compute_cost(train_X_clean, train_y_clean, beta_initial)
initial_cost
```

# IV. Implement Gradient Descent function.

```python
# Define the gradient descent function
def gradient_descent(X, y, beta, learning_rate, num_iterations):
    cost_history = []
    
    for i in range(num_iterations):
        predictions = predict(X, beta)
        errors = predictions - y
        gradient = (1 / len(y)) * np.dot(X.T, errors)
        beta = beta - learning_rate * gradient
        cost = compute_cost(X, y, beta)
        cost_history.append(cost)
        
    return beta, cost_history

# Apply gradient descent
learning_rate = 0.01
num_iterations = 1000
beta_final, cost_history = gradient_descent(train_X_clean, train_y_clean, beta_initial, learning_rate, num_iterations)

# Final coefficients
beta_final
```

# V. Plot the LR model (predicted values) with x-axis and y-axis.
VI. Print coefficients

```python
# Plot the linear regression model
plt.figure(figsize=(10, 6))
plt.scatter(train_clean['LotArea'], train_clean['SalePrice'], color='blue', alpha=0.5, label='Training Data')
plt.plot(train_clean['LotArea'], predict(train_X_clean, beta_final), color='red', label='Fitted Line')
plt.xlabel('LotArea (Normalized)')
plt.ylabel('SalePrice')
plt.title('Linear Regression Model: LotArea vs. SalePrice')
plt.legend()
plt.grid(True)
plt.show()
```

# VI. Print coefficients

```python
# Print final coefficients
print(f'Intercept (β₀): {beta_final[0]:.2f}')
print(f'Slope (β₁): {beta_final[1]:.2f}')
```
